```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import vonmises
import ot
from utils import *

def circular_cost_matrix(n_points):
    """
    Compute the circular cost matrix for points on a unit circle.
    """
    angles = np.linspace(0, 2 * np.pi, n_points, endpoint=False)
    cost_matrix = np.zeros((n_points, n_points))
    
    for i in range(n_points):
        for j in range(n_points):
            distance = min(abs(angles[i] - angles[j]), 2 * np.pi - abs(angles[i] - angles[j]))
            cost_matrix[i, j] = distance**2
    return cost_matrix

def estimate_wasserstein_delta_vonmises(kappas, n_samples=100, n_points=100, n_trials=50, kappa_ls = [1000], eps=2e-3):
    
    """
    Estimate Wasserstein distance between delta measure and von Mises distributions.

    Parameters:
    kappas : list
        Concentration parameters of von Mises distributions.
    n_samples : int
        Number of samples drawn from the von Mises distributions.
    n_points : int
        Number of points to discretize the circle.
    n_trials : int
        Number of trials to estimate mean and standard deviation of Wasserstein distance.

    Returns:
    mean_distances : list
        Mean Wasserstein distances for each kappa.
    std_distances : list
        Standard deviation of Wasserstein distances for each kappa.
    """
    delta_distribution = np.zeros(n_points)
    delta_distribution[0] = 1  # Delta measure at 0 radians
    delta_distribution = ot.unif(n_points)
    angles = np.linspace(0, 2 * np.pi, n_points + 1)
    #cost_matrix = circular_cost_matrix(n_points)
    #cost_matrix_sqrt = cost_matrix#np.sqrt(cost_matrix)
    mean_distances = []
    std_distances = []
    mean_distances_ew = []
    std_distances_ew = []
    for kappa in kappas:
        distances = []
        distances_ew = []
        for _ in range(n_trials):
            np.random.seed(_)
            samples = vonmises.rvs(kappa, size=n_samples)
            histogram, _ = np.histogram(samples % (2 * np.pi), bins=angles, density=True)
            histogram /= histogram.sum()
            distance = ot.emd2(delta_distribution, histogram, cost_matrix)
            distances.append(distance)
            X1 = GM(g=cost_matrix, mode="gauge_only",squared=False, normalize_gauge=False, xi=delta_distribution) # already squared
            X2 = GM(g=cost_matrix, mode="gauge_only",squared=False, normalize_gauge=False, xi=histogram)

            m, ZZ, Z, gamma1, gamma2, zeta1, zeta2, d = perform_grid_search_REW(X1, X2, kappa_ls = kappa_ls, eps=eps, Z_ls=["X2"], 
                                                                                max_len=None)
            #w_nearness = compute_W_nearness(None, zeta1, zeta2, cost_matrix)
            w_nearness = ot.emd2(zeta1, zeta2, torch.from_numpy(cost_matrix))
            #print(ZZ.g.max())
            distances_ew.append(w_nearness)
        mean_distances.append(np.mean(distances))
        std_distances.append(np.std(distances))
        mean_distances_ew.append(np.mean(distances_ew))
        std_distances_ew.append(np.std(distances_ew))

    return mean_distances, std_distances, mean_distances_ew, std_distances_ew

# Parameters
import matplotlib
labelsize = 35
matplotlib.rc('xtick', labelsize=labelsize) 
matplotlib.rc('ytick', labelsize=labelsize)

import matplotlib as mpl
mpl.rcParams['text.usetex'] = True
#mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}'] 
kappas = np.linspace(1, 10, 10, endpoint=True)  # Concentration parameters of von Mises distributions
n_samples = 1_000_000  # Number of samples per distribution
n_points = 360  # Discretization of the circle
n_trials = 1  # Number of trials to estimate distances
eps = 3e-3
kappa_ls = [20]
# Estimate Wasserstein distances
mean_distances, std_distances, mean_distances_ew, std_distances_ew = estimate_wasserstein_delta_vonmises(kappas, n_samples, n_points, n_trials, 
                                                                                                         kappa_ls = kappa_ls, eps=eps)

# Plot results
plt.figure(figsize=(10, 6))
fontsize = 70
#plt.errorbar(kappas, mean_distances, yerr=std_distances, fmt='o-', capsize=5) #, label='Wasserstein Distance')
plt.plot(kappas, mean_distances, marker='o') #, label='Wasserstein Distance')
plt.xlabel('$\mathbf{\kappa}$', fontsize=fontsize, usetex=True)
#plt.gcf().subplots_adjust(right=.5)
plt.ylabel('$\mathbf{\mathrm{\mathbf{EW}}^2_{\mathbf{\cdot}}}$', fontsize=fontsize, usetex=True)
plt.tick_params(axis='y', which='both', labelleft=False, labelright=True)
plt.ylim(0, np.pi)
plt.grid(True)
plt.show()


# Plot results
plt.figure(figsize=(10, 6))
#plt.errorbar(kappas, mean_distances_ew, yerr=std_distances_ew, fmt='o-', capsize=5) #, label='Wasserstein Distance')
plt.plot(kappas, mean_distances_ew, marker='o') #, label='Wasserstein Distance')
plt.xlabel('$\mathbf{\kappa}$', fontsize=fontsize, usetex=True)
plt.ylabel('$\mathbf{\mathrm{\mathbf{EW}}_{\mathbf{\lambda}}^2}$', fontsize=fontsize, usetex=True)
plt.tick_params(axis='y', which='both', labelleft=False, labelright=True)
plt.ylim(0, np.pi)
plt.grid(True)
plt.show()
```
